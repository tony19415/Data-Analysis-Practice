```python
import pandas as pd
```

```python
df = pd.read_csv("RetailSupplyChainData.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
#check for null values
df.isnull().sum()
```

```python
#convert order date and ship date into datetime
df['Order Date2'] = pd.to_datetime(df['Order Date'])
df['Ship Date2'] = pd.to_datetime(df['Ship Date'])
```

```python
#convert the numerical values into int/float
#df['Sales'] = df['Sales'].astype(float)
```

```python
#check data types
df.dtypes
```

```python
filtered = df[['Order ID', 'Order Date2', 'Ship Date2', 'Customer ID', 'Customer Name', 'Country', 'City', 'State', 'Product ID', 'Category', 'Sales', 'Quantity', 'Discount', 'Profit', 'Dates']]
filtered
```

```python
#sort by profit from highest to lowest
filtered.sort_values(['Profit'], ascending=[False])
```

```python
#count entry per cateogry
category_count = filtered['Category'].value_counts()
category_count
```

```python
#count entry per country
country_count = filtered['Country'].value_counts()
country_count
```

```python
#count entry per customer
cust_count = filtered['Customer Name'].value_counts()
cust_count
```

```python
# Calculate the total revenue for each category.

total_revenue_by_cat = filtered.groupby('Category')['Profit'].sum().reset_index()
total_revenue_by_cat.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a bar chart using seaborn
sns.barplot(data = total_revenue_by_cat, x='Category', y='Profit') # Pass column names as strings

# Customize the plot
plt.xlabel("Category")
plt.ylabel("Total Revenue")
plt.title("Total Revenue for Each Category")

# Show the plot
plt.show()
```

```python
# Calculate the total sale for each Product line.
avg_revenue_by_city = filtered.groupby('State')['Sales'].mean().reset_index()

# Create a bar plot using seaborn
sns.barplot(data=avg_revenue_by_city, x='State', y='Sales')

# Add labels and title
plt.xlabel('Product line')
plt.ylabel('Total')
plt.title('Product line vs. Total')

# Rotate x-axis labels by 45 degrees
plt.xticks(rotation=90)

# Show the plot
plt.show()
```
